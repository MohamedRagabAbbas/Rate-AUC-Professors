```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load pre-trained model and tokenizer
model_name = "distilbert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Define input text
text = "He's great. Just keep up with the workload and don't miss lectures"

# Tokenize the input text
inputs = tokenizer(text, return_tensors="pt")

# Perform inference
outputs = model(**inputs)
logits = outputs.logits
predicted_class = logits.argmax().item()

# Interpret the results
sentiment_labels = ["negative", "neutral", "positive"]
predicted_sentiment = sentiment_labels[predicted_class]

# Output the predicted sentiment
print("Predicted Sentiment:", predicted_sentiment)
```

```python
import requests

API_URL = "https://api-inference.huggingface.co/models/SamLowe/roberta-base-go_emotions"
headers = {"Authorization": "Bearer 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"}

def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

# List of input queries
input_queries = [
    "Really good! Definitely go for her. The class could be a bit demanding but you will learn a lot.",
    "I'm feeling quite sad today. Everything seems to be going wrong.",
    "Excited to see my friends after a long time. It's been too long!",
]

# Mapping of sentiment labels to emojis
sentiment_to_emoji = {
    "admiration": "",
    "amusement": "",
    "anger": "",
    "annoyance": "",
    "approval": "",
    "caring": "",
    "confusion": "",
    "curiosity": "",
    "desire": "",
    "disappointment": "",
    "disapproval": "",
    "disgust": "",
    "embarrassment": "",
    "excitement": "",
    "fear": "",
    "gratitude": "",
    "grief": "",
    "joy": "",
    "love": "",
    "nervousness": "",
    "optimism": "",
    "pride": "",
    "realization": "",
    "relief": "",
    "remorse": "",
    "sadness": "",
    "surprise": ""
}


# Output list to store predicted sentiments
predicted_sentiments = []

# Loop over each input query and make a prediction
for query_text in input_queries:
    output = query({"inputs": query_text})
    # Get the predicted sentiment label
    sentiment_label = output[0][0]['label']
    # Map the sentiment label to its corresponding emoji
    emoji = sentiment_to_emoji.get(sentiment_label, "")
    # Append the emoji to the output list
    predicted_sentiments.append(emoji)

# Output the predicted emojis for each input query
for i, emoji in enumerate(predicted_sentiments):
    print(f"Query {i+1} - Predicted Sentiment: {emoji}")
```
